```python
# 01_download_explore.ipynb

# Імпорти
import sys
sys.path.append('..')  # додаємо шлях до src

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.config import DATA_DIR
from src.data_loader import download_yoochoose, load_clicks, load_buys

# Налаштування відображення
%matplotlib inline
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# 1. Завантаження даних
print("="*50)
print("КРОК 1: Завантаження даних")
print("="*50)

download_yoochoose()
clicks = load_clicks()
buys = load_buys()

# 2. Перший погляд на дані
print("\n" + "="*50)
print("КРОК 2: Перший погляд на дані")
print("="*50)

print(f"\nClicks shape: {clicks.shape}")
print(f"Buys shape: {buys.shape}")

print("\nClicks перші 5 рядків:")
display(clicks.head())

print("\nBuys перші 5 рядків:")
display(buys.head())

# 3. Базова інформація
print("\n" + "="*50)
print("КРОК 3: Базова інформація")
print("="*50)

print("\nClicks info:")
print(clicks.info())

print("\nBuys info:")
print(buys.info())

# 4. Статистика
print("\n" + "="*50)
print("КРОК 4: Статистика")
print("="*50)

print("\nClicks статистика:")
display(clicks.describe(include='all'))

print("\nBuys статистика:")
display(buys.describe(include='all'))

# 5. Унікальні значення
print("\n" + "="*50)
print("КРОК 5: Унікальні значення")
print("="*50)

print(f"Унікальних сесій в clicks: {clicks['Session_ID'].nunique():,}")
print(f"Унікальних товарів в clicks: {clicks['Item_ID'].nunique():,}")
print(f"Унікальних сесій в buys: {buys['Session_ID'].nunique():,}")
print(f"Унікальних товарів в buys: {buys['Item_ID'].nunique():,}")

# 6. Часовий аналіз
print("\n" + "="*50)
print("КРОК 6: Часовий аналіз")
print("="*50)

print(f"Період clicks: {clicks['Timestamp'].min()} до {clicks['Timestamp'].max()}")
print(f"Тривалість: {clicks['Timestamp'].max() - clicks['Timestamp'].min()}")

# 7. Візуалізація
print("\n" + "="*50)
print("КРОК 7: Візуалізація")
print("="*50)

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Події по днях
clicks['Date'] = clicks['Timestamp'].dt.date
daily_counts = clicks.groupby('Date').size()
daily_counts.plot(ax=axes[0,0], title='Події по днях', grid=True)
axes[0,0].tick_params(axis='x', rotation=45)

# Події по годинах
clicks['Hour'] = clicks['Timestamp'].dt.hour
hourly_counts = clicks.groupby('Hour').size()
hourly_counts.plot(kind='bar', ax=axes[0,1], title='Події по годинах', grid=True)

# Топ категорій
top_categories = clicks['Category'].value_counts().head(10)
top_categories.plot(kind='bar', ax=axes[1,0], title='Топ-10 категорій', grid=True)

# Співвідношення clicks/buys
axes[1,1].pie([len(clicks), len(buys)], labels=['Clicks', 'Buys'], autopct='%1.1f%%')
axes[1,1].set_title('Співвідношення подій')

plt.tight_layout()
plt.savefig('../figures/eda/initial_overview.png', dpi=100, bbox_inches='tight')
plt.show()

# 8. Збереження інформації
print("\n" + "="*50)
print("КРОК 8: Збереження інформації")
print("="*50)

dataset_info = {
    'total_clicks': len(clicks),
    'total_buys': len(buys),
    'unique_sessions': clicks['Session_ID'].nunique(),
    'unique_items': clicks['Item_ID'].nunique(),
    'unique_categories': clicks['Category'].nunique(),
    'date_range': f"{clicks['Timestamp'].min()} to {clicks['Timestamp'].max()}",
    'clicks_per_session_mean': clicks.groupby('Session_ID').size().mean(),
    'clicks_per_session_median': clicks.groupby('Session_ID').size().median(),
    'conversion_rate': len(buys) / len(clicks) * 100
}

import json
with open(DATA_DIR / 'dataset_info.json', 'w') as f:
    json.dump(dataset_info, f, indent=4, default=str)

print("Інформацію збережено в data/yoochoose/dataset_info.json")
```
